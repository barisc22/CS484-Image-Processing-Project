```python
#Pre-processing
path = 'C:\\Users\\Faruk\\Desktop\\cs484_proje\\data\\train\\'
files = []
labelName = []
for r, d, f in os.walk(path):
    for directory in d:
        labelName.append(directory)


labeledImageList = []
real_labels = []
s = np.array([0.485, 0.456, 0.406 ])
d = np.array([0.229, 0.224, 0.225 ])
i = 0
for f in labelName:
    imageList = []
    for imageName in glob.glob(path + '\\' + f + '\\*.JPEG'):
        Im = Image.open(imageName).convert('RGB')
        temp = np.asarray(Im)
        (height,width,c)= temp.shape
        size = max(height, width)
        #print(size)
        newImage = Image.new('RGB', (size, size), (0,0,0))
        newImage.paste(Im, ((int((size-width)/2), int((size-height)/2))))
        imResized = newImage.resize((224,224))
        image = np.asarray(imResized)
        image = image / 255
        #print(s)
        #print(image)
        image = np.subtract(image, s)
        image = np.divide(image, d)
        image = np.float32(image)
        #print(image)
        imageList.append(image)
        
    labeledImageList.append(imageList)
    i = i + 1

#print(labeledImageList[0][0])
#labeledImageList[0][10](:,:,1) = labeledImageList[0][10](:,:,1)- 0.485
```

```python
for epoch in range(5):  # loop over the dataset multiple times

    running_loss = 0.0
    for i, data in enumerate(my_dataloader, 0):
        # get the inputs; data is a list of [inputs, labels]
        inputs, labels = data

        # zero the parameter gradients
        optimizer.zero_grad()
        #inputs = inputs.squeeze()
        #print(np.shape(inputs))
        #labels = labels.squeeze()
        # forward + backward + optimize
        #print(np.shape(inputs))
        #for i in range(10):
            
        outputs = net(inputs)
        #outputs = outputs.squeeze()
        #print(len(outputs))
        #print(len(labels.squeeze()))
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()

        # print statistics
        running_loss += loss.item()
        if i % 10 == 9:    # print every 2000 mini-batches
            print('[%d, %5d] loss: %.3f' %
                  (epoch + 1, i + 1, running_loss / 10))
            running_loss = 0.0

print('Finished Training')
```
